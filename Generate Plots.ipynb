```python
import matplotlib.pyplot as plt

# Epochs
epochs = [1, 2, 3]

# Accuracy values from your logs
train_acc = [0.6457, 0.6713, 0.6745]
val_acc   = [0.1724, 0.1586, 0.2294]

# Plot
plt.figure()
plt.plot(epochs, train_acc, marker='o', label='Training Accuracy')
plt.plot(epochs, val_acc, marker='o', label='Validation Accuracy')

# Labels and title
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Training vs Validation Accuracy (Warmup Phase)")
plt.xticks(epochs)
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms, models
import os
import numpy as np
from sklearn.utils.class_weight import compute_class_weight
import kagglehub
```

## Download Dataset

```python
path = kagglehub.dataset_download("odins0n/ucf-crime-dataset")
dataset_path = os.path.join(path, 'Train') 
```

## Data Transformations

```python
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
```

## Load Dataset

```python
full_dataset = datasets.ImageFolder(root=dataset_path, transform=transform)
train_size = int(0.8 * len(full_dataset))
val_size = len(full_dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(full_dataset, [train_size, val_size])

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
```

## Handle Imbalance/Compute Class Weights

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

y_train = [full_dataset.targets[i] for i in train_dataset.indices]
class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
weights_tensor = torch.tensor(class_weights, dtype=torch.float).to(device)
```

##  ResNet-34

```python
model = models.resnet34(weights=models.ResNet34_Weights.IMAGENET1K_V1)

#modify the final layer 
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 14) 
model = model.to(device)
```

## Loss and Optimizer

```python
criterion = nn.CrossEntropyLoss(weight=weights_tensor)
optimizer = optim.Adam(model.parameters(), lr=1e-4) #lower lr for fine-tuning
```

## Training and Validation 

```python
def train(model, loader, optimizer, criterion, device):
    model.train()
    running_loss, correct = 0.0, 0
    for images, labels in loader:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        _, predicted = outputs.max(1)
        correct += predicted.eq(labels).sum().item()
    return running_loss/len(loader), correct/len(loader.dataset)

def test(model, loader, criterion, device):
    model.eval()
    running_loss, correct = 0.0, 0
    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            running_loss += loss.item()
            _, predicted = outputs.max(1)
            correct += predicted.eq(labels).sum().item()
    return running_loss/len(loader), correct/len(loader.dataset)
```

```python
EPOCHS = 10
for epoch in range(1, EPOCHS + 1):
    t_loss, t_acc = train(model, train_loader, optimizer, criterion, device)
    v_loss, v_acc = test(model, val_loader, criterion, device)
    print(f"Epoch {epoch} | Train Acc: {t_acc:.4f} | Val Acc: {v_acc:.4f}")
```
